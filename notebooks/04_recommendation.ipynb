```python
import pandas as pd
import numpy as np


from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors


master_df=pd.read_csv(r"C:\Users\HEMANATH\Desktop\FIFA\data\processed\clustered_players.csv")
master_df
print(master_df.columns.tolist())
```

```python
features = [
    "goals_per90",
    "assists_per90",
    "cards_per90",
    "age",
    "log_market_value"
]
```

```python
X=master_df[features].fillna(0)
X
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
X_scaled
```

```python
nn=NearestNeighbors(n_neighbors=6,metric="euclidean")
nn.fit(X_scaled)
```

```python
def recommend_player(player_name):

    matches = master_df[
        master_df["name"]
        .str.lower()
        .str.contains(
            player_name.lower(),
            na=False
        )
    ]

    if len(matches) == 0:
        return "Player not found"

    idx = matches.index[0]

    distances, indices = nn.kneighbors(
        [X_scaled[idx]]
    )

    recommendations = (
        master_df.iloc[indices[0]]
        [
            [
                "name",
                "position_group",
                "age",
                "market_value_in_eur",
                "cluster_name"
            ]
        ]
    )

    return recommendations
```

```python
recommend_player("Haaland")
```

```python
top_scores=master_df.nlargest(20,'goals')[['name','goals','position_group']]
top_scores
```

```python
fig=px.bar(
    top_scores,x='goals',y='name',color='position_group',orientation="h",title="Top 20 Goal Scorers"
)

fig.show()
```

```python
top_assists = (
    master_df
    .nlargest(20, "assists")
    [["name","assists","position_group"]]
)

fig = px.bar(
    top_assists,
    x="assists",
    y="name",
    orientation="h",
    color="position_group",
    title="Top 20 Assist Providers"
)

fig.show()
```

```python
top_value=(master_df.nlargest(20,"market_value_in_eur")[["name","market_value_in_eur","current_club_name"]])

fig=px.bar(top_value,x="market_value_in_eur",y="name",orientation="h",title="top 20 Most Valuable players")
fig.show()
```

```python
fig=px.scatter(master_df,x='age',y='market_value_in_eur',color='position_group',hover_name='name',title='age VS Market Value')
fig.show()
```

```python
fig=px.box(master_df,x="position_group",y='market_value_in_eur',title="Market Value By Postion")
fig.show()
```

```python
fig=px.histogram(master_df,x='age',nbins=30,title="Player Age Distribution")
fig.show()
```

```python
corr_cols=[
    "age",
    "matches",
    "goals",
    "assists",
    "minutes_played",
    "market_value_in_eur",
    "goals_per90",
    "assists_per90",
    "goal_contribution_per90"
]
```

```python
corr_matrix=(master_df[corr_cols].corr(numeric_only=True))
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',fmt=".2f")

plt.title('Correlation Heat MAp')

plt.show()
```

```python
fig=px.scatter(master_df,x="goals",y="market_value_in_eur",color="position_group",hover_name="name",title="Goal Vs Market VAlue")
fig.show()
```

```python
fig=px.scatter(master_df,x='assists',y='market_value_in_eur',color='position_group',hover_name='name',title='Market Value AS Assists')
fig.show()
```

```python
club_value=(master_df.groupby('current_club_name')['market_value_in_eur'].sum().sort_values(ascending=False).head(20).reset_index())

fig=px.bar(club_value,x="market_value_in_eur",y="current_club_name",orientation="h",title="top club by squad value")
fig.show()
```
